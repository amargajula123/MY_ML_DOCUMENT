# Logistic Regression

```python
import seaborn as sns 
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df["species"].unique()
```

```python
df=df[df["species"]!="virginica"]
```

```python
df.species.unique()
```

```python
df["species"]=df.species.map({"setosa":0,"versicolor":1})
```

## independent and dependent features

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
regression=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV

perameters={'C':[0.5,1,2,3,5,6,7,9,10,30,45],'penalty':['l1','l2','elasticnet']}
log_regressor=GridSearchCV(regression,perameters,scoring='accuracy',cv=5)

log_regressor.fit(X,y)
```

```python
log_regressor.best_params_
```

```python
log_regressor.best_score_
```

```python
log_regressor.predict(X_test)
```

```python
prediction_log=log_regressor.predict(X_test)
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='species')
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_recall_curve,roc_auc_score,f1_score
```

```python
confusion_matrix(y_test,prediction_log)
```

```python
f1_score(y_test,prediction_log)
```

```python
precision_recall_curve(y_test,prediction_log)
```

```python
roc_auc_score(y_test,prediction_log)
```

```python
accuracy_score(y_test,prediction_log)
```

# Decision Tree Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df["species"]=df.species.map({"setosa":0,"versicolor":1,"virginica":2})
```

```python
df["species"]
```

```python
df.iloc[:,:-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

### using max_depth and criterion='entropy'

```python
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])


from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
classifier=DecisionTreeClassifier(criterion='gini')
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])


from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
classifier=DecisionTreeClassifier(criterion='gini',max_depth=2)
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])


from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
clf =tree.DecisionTreeClassifier()
clf
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_trein, X_test, y_train, y_test =train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

params={'max_depth':[2,4,6,8,10,12],
        'min_samples_split':[2,3,4],
        'min_samples_leaf' :[1,2]}
clf =tree.DecisionTreeClassifier()

gcv=GridSearchCV(estimator=clf,param_grid=params)
gcv.fit(X_trein,y_train)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
classifier=DecisionTreeRegressor()
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])

from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
classifier=DecisionTreeRegressor(max_depth=2)
classifier.fit(df.iloc[:,:-1],df.iloc[:,-1])


from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_trein, X_test, y_train, y_test =train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

params={'max_depth':[2,4,6,8,10,12],
        'min_samples_split':[2,3,4],
        'min_samples_leaf' :[1,2]}
clf =tree.DecisionTreeRegressor()

gcv=GridSearchCV(estimator=clf,param_grid=params)
gcv.fit(X_trein,y_train)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
prediction_log=gcv.predict(X_test)
prediction_log
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_recall_curve,roc_auc_score,f1_score

confusion_matrix(y_test,prediction_log)
```

```python
accuracy_score(y_test,prediction_log)
```

# Random Forest 

link :https://www.youtube.com/watch?v=ok2s1vV9XW0

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_trein, X_test, y_train, y_test =train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier 
model=RandomForestClassifier(n_estimators=10)
model.fit(X_trein,y_train)
```

```python
model.score(X_test,y_test)
```

```python
pred_log=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
an=confusion_matrix(pred_log,y_test)
an
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(an,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```
